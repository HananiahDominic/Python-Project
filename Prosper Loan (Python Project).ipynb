# Bivariate Exploration

This investigations will maily focus on the following areas: 

The relationship between loan status and interest rate, 
The correlation between recommendation and investment from friends count,
The connection between employment status and loan term.

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='loanstatus', y='borrowerrate')
plt.xlabel('Loan Status')
plt.ylabel('Interest Rate')
plt.title('Loan Status by Interest Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.regplot(data=df, x='recommendations', y='investmentfromfriendscount', color= 'Blue')
plt.title('Recommendation by Investment from friends')
fig = plt.gcf()
fig.set_size_inches(10, 6)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='employmentstatus', y='term')
plt.xlabel('Employment Status')
plt.ylabel('Loan Term')
plt.title('Loan Term by Employment Status')
plt.xticks(rotation=90)
plt.show()
```

# Investigation undertaken and the Relationship Observed 

Loan Status and Interest Rate:The variability in interest rates across different loan statuses suggests that the interest rate has an impact on the probability of loan default or successful repayment.

Recommendation and investment from friends:This examine the relationship between the investment made by friends and the corresponding recommendations and it clearly indicates that recommendation affects investment from friends.

Employment Status and Loan Term:Distinct patterns emerge when examining the relationship between employment status and loan term, indicating that individuals with specific employment statuses tend to choose either shorter or longer loan durations.
